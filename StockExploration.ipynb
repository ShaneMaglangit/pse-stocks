```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv("stock.csv", na_values=np.nan)
df["Sector"] = df["Sector"].apply(lambda x: x.strip() if x.strip() != "None" else np.nan)
df["Subsector"] = df["Subsector"].apply(lambda x: x.strip() if x.strip() != "None" else np.nan)
df["Date"] = df["Date"].apply(lambda x: x.strip())
df.head()
```

```
df[df["Sector"] == ""]["Stock Code"].unique()
```

```
def fill_sectors(source, code, sector, subsector):
  source["Sector"][source["Stock Code"] == code] = np.nan if sector == "None " else sector
  source["Subsector"][source["Stock Code"] == code] = np.nan if subsector == "None " else subsector

df_sectors = pd.read_csv("sector.csv")

for _, sector in df_sectors.iterrows():
  fill_sectors(df, sector[0], sector[1], sector[2])

df.to_csv("stock.csv", index=False)
```
